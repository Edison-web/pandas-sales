List of items in a dictionary

```python

Dict ={
    'product_id': [23, 96, 97, 15, 87],
    'name': ['computer','Python Workout', 'Pandas Workout', 'banana', 'sandwich' ],
    'wholesale_price': [500, 35, 35, 0.5, 3], 
    'retail_price': [1000, 75, 75, 1, 5],
    'sales':[100, 1000, 500, 200, 300]
    }
```

imported needed modules

```python

import numpy as np
import pandas as pd

df = pd.DataFrame(Dict)
df
```

First, I calculated 15% of the original net revenue as net_revenue
Next, I calculated the original net revenue as original_net_revenue
I then deducted the original revenue from the net revenue and named it net_revenue(15%)

```python


net_revenue = ((df['retail_price'] - df['wholesale_price']) * df['sales']) * 0.15
original_net_revenue = (df['retail_price'] - df['wholesale_price']) * df['sales']
df['net_revenue(15%)'] = (original_net_revenue) - (net_revenue)
df
```
